```python
import matplotlib.pyplot as plt
import numpy as np

# Grid setup (N=5 as in your main_script.py)
x = np.linspace(0, 1, 5)

# The analytical solution for the optimized state: u(x) = x^2 - 1.25x + 0.25
# Our solver found w ≈ 0.25, which is the intercept at x=0
u_optimized = x**2 - 1.25*x + 0.25

plt.figure(figsize=(9, 5))
plt.plot(x, u_optimized, 'o-', color='#1f77b4', linewidth=2, markersize=8, label='Optimized State $u(x)$')

# Formatting the plot
plt.title('PDE Optimization Result: State Variable Profile', fontsize=14, fontweight='bold')
plt.xlabel('Spatial Domain ($x$)', fontsize=12)
plt.ylabel('State Value ($u$)', fontsize=12)
plt.axhline(0, color='black', linewidth=0.8)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(frameon=True)

plt.show()
```
